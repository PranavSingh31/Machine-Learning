```python
lss = Lasso(alpha=0.5748)

lss.fit(x_train, y_train)
y_pred_lss = lss.predict(x_test)
print(y_pred_lss[0:5])
print("Lasso Regression R2_Score = ", r2_score(y_test, y_pred_lss))
```
